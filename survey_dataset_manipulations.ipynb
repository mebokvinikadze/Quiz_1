```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read the CSV file

```python
df = pd.read_csv("survey_results_public.csv")
```

# Ensure minimum requirements for the dataset

```python
if df.shape[0] < 100 or df.shape[1] < 10:
    raise ValueError("Dataset does not meet minimum requirements.")
```

# Extract desired rows and columns

```python
desired_columns = ["MainBranch", "EdLevel", "Employment", "Country", "ConvertedCompYearly"]
desired_rows = df[desired_columns].copy()  
desired_rows.iloc[:10, :10]
```

# Drop rows with NaN values in the 'EdLevel' and 'ConvertedCompYearly' columns

```python
desired_rows.dropna(subset=['EdLevel', 'ConvertedCompYearly'], inplace=True)
```

# Assign indexing to a specific column

```python
df.set_index('MainBranch', inplace=True)
df
```

 # Create a filter based on 2 parameters

```python
filter_condition = (desired_rows["Country"] == "United States of America")
filtered_table = desired_rows[filter_condition]
filtered_table
```

# Sort the table using the 2 parameters

```python
df_sorted = df.sort_values(by=['EdLevel', 'ConvertedCompYearly'], ascending=[True, True])
df_sorted
```

# Use statistical functions

```python
mean_salary = df["ConvertedCompYearly"].mean()
std_dev_salary = df["ConvertedCompYearly"].std()
median_salary = df["ConvertedCompYearly"].median()
min_salary = df["ConvertedCompYearly"].min()
max_salary = df["ConvertedCompYearly"].max()

print("Mean Salary:", mean_salary)
print("Standard Deviation of Salary:", std_dev_salary)
print("Median Salary:", median_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
df["Age"].value_counts().sort_index().plot(kind="bar", color="skyblue")
plt.title("Age Distribution of Developers in the US")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Line chart
sopartfreq_counts = df['SOPartFreq'].value_counts()
plt.figure(figsize=(12, 6))
sopartfreq_counts.plot(kind='line', marker='o', color='green', linestyle='-', linewidth=2, markersize=8)
plt.title('Distribution of Stack Overflow Participation Frequency')
plt.xlabel('Frequency of Participation')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
